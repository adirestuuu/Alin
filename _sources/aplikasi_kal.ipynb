
#**Page Rank**

PageRank (PR) merupakan sebuah metrik yang dikembangkan oleh Google untuk mengukur tingkat kepentingan sebuah halaman web. Konsep dasarnya sederhana: semakin banyak halaman lain yang menautkan ke halaman tersebut, maka semakin tinggi pula skor PageRank-nya.

**Algoritma PageRank**

PageRank adalah algoritma yang menghitung probabilitas seseorang tiba di halaman tertentu saat mengklik tautan secara acak. Perhitungan ini berlaku untuk kumpulan dokumen dengan ukuran berapa pun. Nilai PageRank awal diasumsikan sama untuk semua halaman, kemudian disesuaikan melalui iterasi agar mendekati nilai sebenarnya.

```python

import numpy as np

def hitung_ranking(link_matrix, faktor_redaman=0.85, batas_error=0.000001, max_ulang=100):
    jumlah_halaman = len(link_matrix)
    ranking = np.full(jumlah_halaman, 1/jumlah_halaman)

    for _ in range(max_ulang):
        ranking_lama = ranking.copy()
        ranking = (1 - faktor_redaman) / jumlah_halaman + faktor_redaman * np.dot(link_matrix.T, ranking)

        if np.sum(np.abs(ranking - ranking_lama)) < batas_error:
            break

    return ranking

# Matriks link antar halaman
link_matrix = np.array([
    [0, 0, 1, 0.5],
    [1/3, 0, 0, 0],
    [1/3, 0.5, 0, 0.5],
    [1/3, 0.5, 0, 0]
])

# Hitung dan cetak hasilnya
hasil_ranking = hitung_ranking(link_matrix)
print("Skor PageRank:", hasil_ranking)
```
